```python
# TODO собрать импорты всех модулей в одном месте
import pandas as pd
import geopandas as gpd
import osmnx as ox
from keplergl import KeplerGl
```

```python
TILES = "CartoDB positron"  # Название подложки для карт

# TODO указать любой район Санкт-Петербург из OSM https://wiki.openstreetmap.org/wiki/RU:%D0%A1%D0%B0%D0%BD%D0%BA%D1%82-%D0%9F%D0%B5%D1%82%D0%B5%D1%80%D0%B1%D1%83%D1%80%D0%B3/%D0%A0%D0%B0%D0%B9%D0%BE%D0%BD%D1%8B
TERRITORY_NAME = 'Выборгский район, Санкт-Петербург'  # название территории для которой будут строиться слои

# TODO указать ссылку на файл из вашего github репозитория, которая начинается с https://raw.githubusercontent.com/
KGIOP_FILE_URL = "https://raw.githubusercontent.com/lbvvnvn/PythonITMO_2023/main/GeoJSON_Layers/kgiop_objects.geojson"  # ссылка на слой с объектами культурного наследия
STREETS_FILE_URL = "https://raw.githubusercontent.com/lbvvnvn/PythonITMO_2023/main/GeoJSON_Layers/streets.geojson"  # ссылка на слой с улицами
```

## Территория

### Загрузка территории из OSM (Extract)

```python
# TODO загрузить geodataframe с геометрией для территории TERRITORY_NAME
district = ox.geocode_to_gdf(TERRITORY_NAME)
district
```

```python
district.explore(tiles=TILES)
```

## Улицы

### Загрузка файла с улицами (Extract)

```python
# TODO отфильтровать улицы по маске геометрии территории полученной ранее
streets = gpd.read_file(STREETS_FILE_URL, mask=district)
streets
```

### Обработка данных с улицами (Transform)

```python
# TODO сгруппировать и объединить геометрии с одинаковыми названиями
streets = streets.dissolve(by="name")
streets
```

```python
streets.explore(tiles=TILES)
```

### Сохранение слоя с улицами (Load)

```python
# TODO переименовать столбцы в русские названия, кроме столбца geometry

# TODO для того чтобы переименовать индекс, нужно обратиться и нему и вызвать от него метод rename (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Index.rename.html)
streets.index.rename('Название', inplace= True)

streets.head()
```

*После объединения с момощью метода dissolve, не устраивает результат того, что удаляются индексы. Вместо них выступают уникальные значениям улиц.
Вопрос: можно ли объединять геометрию таким образом, чтобы индексы обновлялись, а названия улиц оставались в формате колонки?*

```python
streets.reset_index(inplace=True)
streets.head()
```

```python
# TODO сохранить слой в географической проекции в формате GeoJSON
streets.crs
```

```python
streets.to_crs(4326).to_file('streets_wgs.geojson', driver='GeoJSON')
```

```python
from google.colab import output
output.enable_custom_widget_manager()
```

```python
map_streets = KeplerGl(height=500)
map_streets.add_data(data=streets.copy(), name="Улицы")  # на всякий делаем копию :)

map_streets
```

```python
map_streets.save_to_html()
```

## Объекты культурного наследия

### Загрузка объектов культурного наследия (Extract)

```python
# TODO отфильтровать улицы по маске геометрии территории полученной ранее
okn = gpd.read_file(KGIOP_FILE_URL, mask=district)
okn
```

```python
okn.info()
```

### Обработка объектов культурного наследия (Transform)

```python
okn.drop(columns="id", inplace=True)  # удаляет указанные столбцы
okn.head()
```

```python
okn.duplicated().value_counts()
```

```python
# TODO добавить два столбца lon и lat, в которых будут долгота и широта
okn["lon"] = okn.to_crs(3857).geometry.centroid.to_crs(4326).x
okn["lat"] = okn.to_crs(3857).geometry.centroid.to_crs(4326).y
okn
```

### Сохранение слоя с объектами культурного наследия (Load)

```python
# TODO переименовать столбцы в русские названия, кроме столбца geometry
rename_columns = {
    "ensemble_name": "Название ансамбля",
    "object_name": "Название объекта",
    "occurrence_time": "Время возникновения",
    "object_location": "Адрес",
    "historical_category": "Категория ИКЗ ОКН",
    "normative_act": "Нормативный акт",
    "object_type": "Вид ОКН",
    "lon": "Долгота (х)",
    "lat": "Широта (у)"
}    # Перевод значений некоторых позиций осуществлен в соответсвии с Федеральным законом N 73-ФЗ "Об объектах культурного наследия..."
okn.rename(columns=rename_columns, inplace=True)
```

```python
okn.head()
```

```python
# TODO сохранить слой в географической проекции в формате GeoJSON
okn.to_crs(4326).to_file('okn_wgs.geojson', driver='GeoJSON')
```

```python
map_okn = KeplerGl(height=500)
map_okn.add_data(data=okn.copy(), name="Объекты культурного")  # на всякий делаем копию :)

map_okn
```
